```python
class Solution:
    def insert(self, intervals: list[list[int]], newInterval: list[int]) -> list[list[int]]:
        i = 0
        n = len(intervals)
        result = []
        while (i < n) and (intervals[i][1] < newInterval[0]):
            result.append(intervals[i])
            #return result
            i += 1
            
        while (i < n) and (newInterval[0] <= intervals[i][1]) and (newInterval[1] >= intervals[i][0]):
            newInterval = [min(intervals[i][0], newInterval[0]), max(intervals[i][1], newInterval[1])]
            i += 1
        result.append(newInterval)

        while (i < n) :
            result.append(intervals[i])
            i += 1
        return result

intervals = [[3,5],[6,7],[8,10],[12,16]]
newInterval = [17, 18]
Solution().insert(intervals,newInterval)
```

```python
class Solution:
    def insert(self, intervals: list[list[int]], newInterval: list[int]) -> list[list[int]]:
        
        result = []
        i = 0 # 使用索引來追蹤當前位置
        N = len(intervals)
        
        # --- 階段一：處理左側不重疊區間 ---
        # 條件：舊區間 i 完全在 newInterval 左側 (end_i < start_new)
        while i < N and intervals[i][1] < newInterval[0]:
            result.append(intervals[i])
            i += 1
            
        # --- 階段二：處理重疊並合併 ---
        # 條件：只要有重疊就持續合併 (start_i <= end_new)
        while i < N and intervals[i][0] <= newInterval[1]:
            # 擴大 newInterval 的範圍
            # newInterval[0] 更新為 min(當前舊區間的起始, newInterval 的起始)
            newInterval[0] = min(intervals[i][0], newInterval[0])
            # newInterval[1] 更新為 max(當前舊區間的結束, newInterval 的結束)
            newInterval[1] = max(intervals[i][1], newInterval[1])
            i += 1
            
        # 階段二結束後，將合併完成的 newInterval 加入結果列表
        result.append(newInterval)
        
        # --- 階段三：處理右側不重疊區間 ---
        # 條件：剩下的區間都位於 newInterval 的右側，且不重疊
        # 由於是 while 迴圈，i 已經停在第一個右側不重疊的區間上了
        while i < N:
            result.append(intervals[i])
            i += 1
            
        return result
```
