```python
# 寫一個函數，找出字串陣列中最長的公共前綴字串。

# 如果沒有公共前綴，則傳回空字串""。

# 例1：
# 輸入： strs = ["flower","flow","flight"]
# 輸出： "fl"

# 例2：
# 輸入： strs = ["dog","racecar","car"]
# 輸出： ""
# 說明：輸入字串之間沒有公共前綴。
 

# 限制條件：
# 1 <= strs.length <= 200
# 0 <= strs[i].length <= 200
# strs[i]如果非空，則僅包含小寫英文字母。

#訣竅:先定住一個str再去比其他的str
```

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        for i, in strs:
            strs[i][j] 
```

```python
strs = ["flower","flow","flight"]
# 處理空數組的情況 (雖然限制條件說 strs.length >= 1, 但養成習慣是好的)
if not strs:
    print("")

# 選擇第一個字串作為比較的基準
first_str = strs[0]

# 1. 遍歷第一個字串的每一個字符（決定前綴的長度）
for j in range(len(first_str)):
    # 獲取基準字符
    char = first_str[j]
    
    # 2. 將此字符與數組中所有其他字串進行對比 (垂直掃描)
    # 從索引 i=1 開始，遍歷到數組的末尾
    for i in range(1, len(strs)):
        # 3. 檢查兩個停止條件
        
        # 條件 A: 索引越界 (當前字串 strs[i] 比基準字串 strs[0] 短)
        # 或是
        # 條件 B: 字符不匹配 (strs[i][j] 不等於基準字符 char)
        if j == len(strs[i]) or strs[i][j] != char:
            # 一旦發現不匹配或長度不夠，最長公共前綴就是
            # 基準字串從頭到當前索引 j 之前的部分。
            return first_str[:j]

# 4. 如果第一個字串被完全遍歷完畢都沒有觸發停止條件，
#    則表示第一個字串本身就是最長公共前綴（例如 strs=["a", "a", "a"] 或 strs=["flower", "flower"]）
return first_str
```
