```python
# 寫一個函數，找出字串陣列中最長的公共前綴字串。

# 如果沒有公共前綴，則傳回空字串""。

# 例1：
# 輸入： strs = ["flower","flow","flight"]
# 輸出： "fl"

# 例2：
# 輸入： strs = ["dog","racecar","car"]
# 輸出： ""
# 說明：輸入字串之間沒有公共前綴。
 

# 限制條件：
# 1 <= strs.length <= 200
# 0 <= strs[i].length <= 200
# strs[i]如果非空，則僅包含小寫英文字母。

#訣竅:先定住一個str再去比其他的str
```

```python
from typing import List
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        if not strs:
            return ''

        first_str = strs[0]
        for j in range(len(first_str)): #=strs[o][j],輸出為'f'
            for i in range(1, len(strs)):  #從1開始是因為前面已用strs[0]=first_str來做比較，從每個str的第一字元比較
        
                if j >= len(strs[i])  or first_str[j] != strs[i][j]:
                    return first_str[:j]

        return first_str

# 測試
strs = ["flower", "flow", "flight"]
print(f"輸入: {strs}")
print(f"輸出: \"{Solution().longestCommonPrefix(strs)}\"")

strs_short = ["a", "b"]
print(f"輸入: {strs_short}")
print(f"輸出: \"{Solution().longestCommonPrefix(strs_short)}\"")
```

```python
#優化參考+上zip()方法
from typing import List

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        # 1. (修正) 處理空數組的情況並立即返回
        if not strs:
            return ""
        
        # 2. (優化) 使用 zip 進行垂直掃描
        # zip(*strs) 會將 ['flower', 'flow', 'flight'] 轉換為
        # [('f', 'f', 'f'), ('l', 'l', 'l'), ('o', 'o', 'i'), ...]
        
        prefix = ""
        
        # 遍歷由垂直字符組成的元組
        for chars in zip(*strs):
            # 檢查元組中的所有字符是否相同
            # set(chars) 只有一個元素，說明所有字符都一樣
            if len(set(chars)) == 1:
                # 如果相同，將該字符添加到前綴中
                prefix += chars[0]
            else:
                # 只要發現不匹配，立即返回當前前綴
                return prefix
        
        # 如果所有字符都被遍歷完（意味著所有字串都一樣長且內容相同，或已達到最短字串的長度）
        return prefix

# 測試
strs = ["flower", "flow", "flight"]
print(f"輸入: {strs}")
print(f"輸出: \"{Solution().longestCommonPrefix(strs)}\"")

strs_short = ["a", "b"]
print(f"輸入: {strs_short}")
print(f"輸出: \"{Solution().longestCommonPrefix(strs_short)}\"")
```
