```python
# 寫一個函數，找出字串陣列中最長的公共前綴字串。

# 如果沒有公共前綴，則傳回空字串""。

# 例1：
# 輸入： strs = ["flower","flow","flight"]
# 輸出： "fl"

# 例2：
# 輸入： strs = ["dog","racecar","car"]
# 輸出： ""
# 說明：輸入字串之間沒有公共前綴。
 

# 限制條件：
# 1 <= strs.length <= 200
# 0 <= strs[i].length <= 200
# strs[i]如果非空，則僅包含小寫英文字母。

#訣竅:先定住一個str再去比其他的str
```

```python
class Solution:
    from typing import List
    def longestCommonPrefix(self, strs: List[str]) -> str:
        if not strs:
            print('')

        first_str = strs[0]
        for j in range(len(first_str)): #=strs[o][j],輸出為'f'
            for i in range(1, len(strs)):  #從1開始是因為前面已用strs[0]=first_str來做比較，從每個str的第一字元比較
        
                if j >= len(strs[i])  or first_str[j] != strs[i][j]:
                    return first_str[:j]

        return first_str

strs = ["flower","flow","flight"]
print(Solution().longestCommonPrefix(strs))
```
