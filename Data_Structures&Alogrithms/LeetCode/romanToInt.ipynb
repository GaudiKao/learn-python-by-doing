```python
# 羅馬數字以七個不同的符號表示：  Ia、Vb、Xc、Ld、Ce、iD和b M。
# 符號:值 = {'I' : 1, 'V' : 5, 'X' : 10, 'L' : 50, 'C' : 100, 'D' : 500, 'M' : 1000}

# 例如，用羅馬數字2表示為II ，即兩個 1 相加。 用羅馬數字12表示為 XII。即X + II數字。 127以羅馬數字表示為 XXVII，即 XX + V + II。
# 要先了解羅馬數字的組合規則
# 羅馬數字通常從左到右按從大到小的順序書寫。然而，數字 4 的寫法並非IIII如此。它寫作 4。IV因為 1 在 5 前面，所以我們減 5 得到 4。同樣的道理也適用於數字 9，它的寫法是 IX ，共有六種情況下需要用到減法：

# I可以放在V（5）和X（10）之前，組成4和9。 
# X可以放在L（50）和C（100）之前，使之變為40和90。 
# C可以放在D（500）和M（1000）之前，使之變為400和900。
# 給定一個羅馬數字，將其轉換為整數。

# 例1：
# 輸入： s = "III"
# 輸出： 3
# 說明： III = 3。

# 例2：
# 輸入： s = "LVIII"
# 輸出： 58
# 說明： L = 50，V = 5，III = 3。

# 例3：
# 輸入： s = "MCMXCIV"  只有I、X、C可以做為小數在左邊減去 且 左邊小數只能比右邊大一位數
# 輸出： 1994
# 說明： M = 1000，CM = 900，XC = 90，IV = 4。

# 限制條件：
# 1 <= s.length <= 15
# s僅包含字元('I', 'V', 'X', 'L', 'C', 'D', 'M')。
# 可以保證 這s是一個有效的羅馬數字，其範圍在 0 到 1 之間[1, 3999]。
```

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        roman_table = {'I' : 1, 'V' : 5, 'X' : 10, 'L' : 50, 'C' : 100, 'D' : 500, 'M' : 1000}
        previout_value = roman_table[s[-1]]
        for i in range(len(s)-2, -1, -1):
            current_value = roman_table[s[i]]
            if current_value < previout_value:  #減法情況
                previout_value -= current_value
            else:                               #加法情況
                previout_value += current_value 
        return previout_value
    
s = 'MCMXCIV'
print(Solution().romanToInt(s))
```

```python
#參考對照
table = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}  # 轉換對照表
roman = [i for i in input()]             # 將輸入的羅馬數字變成串列
r = roman[::-1]                          # 反轉串列
output = table[r[0]]                     # 讓 output 先等於第一個數字
for i in range(1, len(r)):               # 從第二個數字開始依序取到最後一個數字
    if table[r[i]] < table[r[i-1]]:      # 如果後面數字比較小
        output = output - table[r[i]]    # 讓 output 減去後面的數字
    else:
        output = output + table[r[i]]    # 如果後面數字比較大，讓 output 加上後面的數字
print(output)
```
