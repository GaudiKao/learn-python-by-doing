```python
# 給定一個整數x，true如果x是一個整數則回傳。回文以及false其他情況。
# 例1：
# 輸入： x = 121
# 輸出： true
# 解釋： 121 從左到右讀作 121，從右到左讀作 121。

# 例2：
# 輸入： x = -121
# 輸出： false

# 例3：
# 輸入： x = 10
# 輸出： false
# 解釋：從右到左讀作 01。因此，它不是回文數。       
```

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        x =  str(x)
        for i in range(len(x)):
            if x == x[::-1]:  #x[start:stop:step]
                return 'true'
            else:
                return 'false'

x = -2222
print(type(x))
print(Solution().isPalindrome(x))
```
